```python
import os
import torch
import numpy as np

from transformers import BloomTokenizerFast, BloomForCausalLM
```

```python
ROOT_PATH = '/content/drive/MyDrive/Colab Notebooks'
PROJ_PATH = os.path.join(ROOT_PATH, 'PyTorch')
DATA_PATH = os.path.join(ROOT_PATH, 'data')
MODEL_ROOT_PATH = os.path.join(PROJ_PATH, 'model')
```

```python
tokenizer = BloomTokenizerFast.from_pretrained("bigscience/bloom-1b7")
```

```python
model = BloomForCausalLM.from_pretrained("bigscience/bloom-1b7")
```

```python
MODEL_PATH = os.path.join(MODEL_ROOT_PATH, 'bloom-1b7')
```

```python
model.save_pretrained(MODEL_PATH)
```

```python
model = BloomForCausalLM.from_pretrained(MODEL_PATH)
```

```python
additional_text_length = 400
starting_text = """Write an email elaborating on all of the following points. 
Sender: Paul, the Head of IT
Receiver: John, a colleague at work
- Thank him for the call this afternoon
- Say you are more familiar with the restructuring planned for next Quarter
Ask him to do the following:
- Book a meeting at 4pm next Wednesday
- Review the attached legal contract by tomorrow
The email should be 3 paragraphs long and elaborate on the above points. 

Email:
"""
```

```python
inputs = tokenizer.encode(starting_text, return_tensors="pt")
print(inputs)
```

```python
predicted_ids = model.generate(
    inputs,
    max_length = len(inputs[0]) + additional_text_length,
)
print(predicted_ids)
```

```python
print(tokenizer.decode(predicted_ids[0]))
```
